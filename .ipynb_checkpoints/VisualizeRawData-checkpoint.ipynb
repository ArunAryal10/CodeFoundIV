# load actual betas

```python
def LoadActualBetas(subj, input_network):
    """
    Loads trial-wise beta activations for left and right button presses
    from HDF5 files, grouped by the specified input network.

    Parameters
    ----------
    subj : str
        Subject identifier (used to locate the subject-specific .h5 file).
    input_network : int
        Index of the target network in the 'networkdef' array. Used to select relevant parcels.

    Returns
    -------
    ActualActivationVecLeft : array or None, shape (N, P)
        Beta activations for N left-press trials across P parcels in the input network.
    ActualActivationVecRight : array or None, shape (M, P)
        Beta activations for M right-press trials across P parcels in the input network.
    """
    ActualActivationVecLeft = []
    ActualActivationVecRight = []

    h5_path = actual_data_dir + subj + actual_suffix  # Full path to subject's .h5 file

    try:
        with h5py.File(h5_path, 'r') as h5f:
            for run in allRuns:
                # Skip missing runs
                if run not in h5f:
                    continue

                for miniblock_key in h5f[run]:
                    mb = h5f[run][miniblock_key]

                    # Skip miniblocks without motor response info
                    if 'motor_response' not in mb:
                        continue

                    # Load all conditions, betas, and motor responses for the miniblock
                    conditions = [c.decode('utf-8') for c in mb['condition']]
                    betas = mb['betas'][:]  # shape: (nTrials, nParcels)
                    motor_resp = [r.decode('utf-8') for r in mb['motor_response']]

                    # Loop over trials in this miniblock
                    for i, cond in enumerate(conditions):
                        if 'GO' in cond:
                            goBetas = betas[i].astype(np.float32, copy=False)
                            button = motor_resp[i]

                            # Validate dimensionality
                            if goBetas.shape[0] != len(networkdef):
                                raise ValueError("Mismatch between GO betas and networkdef dimensions.")

                            # Select only parcels belonging to the input network
                            input_mask = (networkdef == input_network)

                            # Append the beta vector to left/right list based on motor response
                            if button == 'left':
                                actVectLeft = goBetas[input_mask]
                                if actVectLeft.size > 0:
                                    ActualActivationVecLeft.append(actVectLeft)

                            elif button == 'right':
                                actVectRight = goBetas[input_mask]
                                if actVectRight.size > 0:
                                    ActualActivationVecRight.append(actVectRight)

    except Exception as e:
        logging.error(f"Failed to process subject {subj} in LoadActualBetas: {e}")

    # Stack trial-wise activations into (nTrials, nParcels) or return None if empty
    ActualActivationVecLeft = np.vstack(ActualActivationVecLeft) if ActualActivationVecLeft else None
    ActualActivationVecRight = np.vstack(ActualActivationVecRight) if ActualActivationVecRight else None

    return ActualActivationVecLeft, ActualActivationVecRight
```

# Group level networkwise PCA 

```python
def performPCA(input_network):
    """Performs PCA on group-averaged left and right beta values for a specified network.

    Parameters
    ----------
    input_network : int
        Index of the target network (1-based, consistent with networkmappings).

    Returns
    -------
    None
        Displays bar plot of variance explained and 3D PCA scatter plot.
    """
    avg_left_all = []
    avg_right_all = []

    # Loop through all subjects to load and average left/right betas
    for subj in subjNums:
        left_betas, right_betas = LoadActualBetas(subj, input_network)

        if left_betas is None or right_betas is None:
            continue
        if len(left_betas) == 0 or len(right_betas) == 0:
            continue

        avg_left_all.append(np.mean(left_betas, axis=0))
        avg_right_all.append(np.mean(right_betas, axis=0))

    # Combine into group matrix and labels
    X_group = np.vstack([avg_left_all, avg_right_all])  # shape: (nSubjects*2, nRegions)
    y_group = np.array([0]*len(avg_left_all) + [1]*len(avg_right_all))  # 0=left, 1=right

    # Run PCA
    pca_group = PCA(n_components=3)
    X_group_pca = pca_group.fit_transform(X_group)

    # Bar plot: variance explained
    plt.figure(figsize=(3, 3))
    plt.bar(range(1, 4), pca_group.explained_variance_ratio_ * 100)
    plt.xticks([1, 2, 3], ['PC1', 'PC2', 'PC3'])
    plt.ylabel('% Variance Explained')
    plt.title(f'Explained Variance for {list(networkmappings.keys())[input_network - 1]}')
    plt.show()

    # 3D scatter plot: PC space
    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(X_group_pca[y_group == 0, 0], X_group_pca[y_group == 0, 1], X_group_pca[y_group == 0, 2],
               label='Left Avg', alpha=0.8)
    ax.scatter(X_group_pca[y_group == 1, 0], X_group_pca[y_group == 1, 1], X_group_pca[y_group == 1, 2],
               label='Right Avg', alpha=0.8)

    ax.set_title(f'Group-Level PCA: {list(networkmappings.keys())[input_network - 1]}')
    ax.set_xlabel('PC1')
    ax.set_ylabel('PC2')
    ax.set_zlabel('PC3')
    ax.legend()
    plt.tight_layout()
    plt.show()
```

```python
# run for all networks
for network in target_networks:
    performPCA(input_network = network)
    
```

# Actual betas decoding (ground truth)

```python
def CVcorr(subj, input_network):
    """
    Performs cross-validated classification using correlation-based template matching.
    Trains on a subset of trials and classifies left/right motor responses.

    Parameters
    ----------
    subj : str
        Subject identifier.
    input_network : int
        Network index for selecting relevant parcel activations.

    Returns
    -------
    avg_acc : float
        Average classification accuracy across cross-validation folds.
    """
    # Load trial-wise actual activations for left and right button presses
    actual_left, actual_right = LoadActualBetas(subj, input_network)
    
    # Combine left and right activations into one matrix (X)
    X = np.concatenate((actual_left, actual_right), axis=0)
    
    # Create label array: 'L' for left trials, 'R' for right trials
    y = np.array(['L'] * len(actual_left) + ['R'] * len(actual_right))
    
    # Initialize 5-fold cross-validation
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    cv_scores = []
    correct_predictions_total = 0
    total_samples = 0

    # Loop over folds
    for train_index, test_index in kf.split(X):
        # Split betas and labels into training and test sets
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        # Compute mean activation template for left and right classes
        template_left = np.mean(X_train[y_train == 'L'], axis=0)
        template_right = np.mean(X_train[y_train == 'R'], axis=0)

        # Classify test set based on which template has higher correlation
        decode = []
        for sample in range(len(X_test)):
            corr_left = np.corrcoef(X_test[sample], template_left)[0, 1]
            corr_right = np.corrcoef(X_test[sample], template_right)[0, 1]
            decode.append('L' if corr_left > corr_right else 'R')

        # Compute and store accuracy for this fold
        correct_predictions = np.sum(np.array(decode) == y_test)
        fold_accuracy = correct_predictions / len(y_test)
        cv_scores.append(fold_accuracy)

        correct_predictions_total += correct_predictions
        total_samples += len(y_test)

    # Average accuracy across all folds
    avg_acc = np.mean(cv_scores)
    return avg_acc


def process_network(subj, input_network):
    """
    Wrapper to compute classification accuracy for one network.
    """
    return CVcorr(subj, input_network)


def process_subject(subj):
    """
    Computes classification accuracy for all networks for a given subject.
    
    Parameters
    ----------
    subj : str
        Subject identifier.

    Returns
    -------
    network_wise_subj : list of float
        Accuracy for each network.
    """
    network_wise_subj = Parallel(n_jobs=-1)(
        delayed(process_network)(subj, input_network) 
        for input_network in target_networks
    )
    print(f"Finished running subject: {subj}")
    return network_wise_subj


# Run the classification pipeline for all subjects and networks in parallel
subject_network_accuracy = Parallel(n_jobs=-1)(
    delayed(process_subject)(subj) for subj in subjNums
)
```

# Bar plot of ground truth grouped by network

```python
# 1. Prepare accuracy data
subject_network_accuracy = np.array(subject_network_accuracy)
subject_network_accuracy = np.clip(subject_network_accuracy, 0, 1)  # Ensure values are within [0, 1]

# Get network labels and color palette
networks = list(networkmappings.items())
network_labels = [name.upper() for name, _ in networks]
networkpalette = [
    'royalblue', 'slateblue', 'paleturquoise', 'darkorchid', 'limegreen',
    'lightseagreen', 'yellow', 'orchid', 'red', 'peru', 'orange', 'olivedrab'
]
assert len(network_labels) == len(networkpalette), "Mismatch between network labels and palette."

# 2. Perform Wilcoxon signed-rank test against chance (0.5)
chance_level = 0.5
t_stat_netwise, p_val_netwise = [], []

for i in range(subject_network_accuracy.shape[1]):
    network_accuracies = subject_network_accuracy[:, i]
    valid = ~np.isnan(network_accuracies)
    
    if not np.any(valid):
        t_stat_netwise.append(np.nan)
        p_val_netwise.append(1.0)
        continue

    differences = network_accuracies[valid] - chance_level
    stat, p_value = wilcoxon(differences)
    t_stat_netwise.append(stat)
    p_val_netwise.append(p_value)

# 3. Apply Bonferroni correction
_, p_val_corrected, _, _ = multipletests(p_val_netwise, method='bonferroni')

# 4. Organize data into long format for plotting
df_accuracy = pd.DataFrame(subject_network_accuracy, columns=network_labels)
df_accuracy['Subject'] = np.arange(1, subject_network_accuracy.shape[0] + 1)
df_melted = df_accuracy.melt(id_vars='Subject', var_name='Network', value_name='Accuracy')

# 5. Map corrected p-values to networks
network_p_values_corrected = dict(zip(network_labels, p_val_corrected))
df_melted['p-value'] = df_melted['Network'].map(network_p_values_corrected)

# Check for missing p-values
if df_melted['p-value'].isnull().any():
    missing_networks = df_melted[df_melted['p-value'].isnull()]['Network'].unique()
    print(f"Warning: Missing p-values for networks: {missing_networks}")
    df_melted['p-value'] = df_melted['p-value'].fillna(1)

# 6. Significance star mapping function
def significance_stars(p):
    if p < 0.001:
        return '***'
    elif p < 0.01:
        return '**'
    elif p < 0.05:
        return '*'
    return ''

df_melted['Significance'] = df_melted['p-value'].apply(significance_stars)

# 7. Compute group statistics
mean_accuracy_per_network = df_melted.groupby('Network')['Accuracy'].mean()
sem_accuracy_per_network = df_melted.groupby('Network')['Accuracy'].sem()
max_accuracy_per_network = df_melted.groupby('Network')['Accuracy'].max()

# 8. Visualization
sns.set_style("white")
plt.figure(figsize=(5, 6))

# Bar plot with SEM
ax = sns.barplot(
    x='Network', y='Accuracy', data=df_melted,
    palette=networkpalette, edgecolor=None,
    errorbar='se', capsize=0.1, errcolor='black', errwidth=1
)

# Add individual data points
sns.stripplot(
    x='Network', y='Accuracy', data=df_melted,
    color='black', alpha=0.5, jitter=False, size=1.5
)

# Add significance stars
for i, network in enumerate(network_labels):
    stars = significance_stars(network_p_values_corrected[network])
    if stars:
        y_max = max_accuracy_per_network[network]
        y_offset = 0.25 * (df_melted['Accuracy'].max() - df_melted['Accuracy'].min())
        ax.text(i, y_max + y_offset, stars, ha='center', va='bottom',
                color='#DC143C', fontsize=16, fontweight='bold')

# Final plot formatting
plt.axhline(y=0.5, color='black', linestyle='--', linewidth=1, label='Chance Level (0.5)')
ax.set_xticklabels(network_labels, fontsize=16, rotation=90)
plt.xlabel('Networks', fontsize=16)
plt.ylabel('Accuracy', fontsize=16)
ax.tick_params(axis='y', labelsize=16)
ax.yaxis.grid(False)
ax.set_ylim(0.3, 0.85)
plt.legend()
plt.tight_layout()
plt.show()
```
